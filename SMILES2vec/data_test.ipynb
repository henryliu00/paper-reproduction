```python
from rdkit import Chem
import rdkit.Chem.AllChem as AllChem
import rdkit.Chem.Descriptors as Descriptors
import rdkit.Chem.rdMolDescriptors as rdMolDescriptors
import rdkit.Chem.EState as EState
import rdkit.Chem.rdPartialCharges as rdPartialCharges
import rdkit.Chem.rdChemReactions as rdRxns
import numpy as np
import pandas as pd

atom_set = set()
atom_list = []

for smile in fs_smiles:
    m = Chem.MolFromSmiles(smile)
    for i in m.GetAtoms():
        atom_set.add( i.GetAtomicNum() )
        atom_list.append( i.GetAtomicNum() )
```

```python
for i in atom_set:
    print(i,':',atom_list.count(i)/len(atom_list))
```

### draw the molecular graph from SMILES (using RDKit)

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
```

```python
fs_mol = [Chem.MolFromSmiles(x) for x in fs_smiles]
Draw.MolsToGridImage(fs_mol[60:70], molsPerRow=5, useSVG=False, legends=fs_smiles[10:20])
```
